```python
import pandas as pd
import os

# ---------------------------------------------------------
# INPUT + OUTPUT LOCATIONS
# ---------------------------------------------------------
input_csv = r"C:\Users\TUF\Desktop\Architecture\UCL\Design\General_Flickr_Scrapping\Castello\Combined_Castello.csv"
output_folder = r"C:\Users\TUF\Desktop\Architecture\UCL\Design\General_Flickr_Scrapping\Castello\By_Year"

# Ensure output directory exists
os.makedirs(output_folder, exist_ok=True)

# ---------------------------------------------------------
# LOAD CSV
# ---------------------------------------------------------
df = pd.read_csv(input_csv, keep_default_na=False)

# ---------------------------------------------------------
# CLEAN DATE COLUMN + EXTRACT YEAR
# ---------------------------------------------------------
df["date_taken"] = pd.to_datetime(df["date_taken"], errors="coerce")

# Extract year
df["year"] = df["date_taken"].dt.year

# Remove rows that don't have a valid year
df = df.dropna(subset=["year"])
df["year"] = df["year"].astype(int)

# ---------------------------------------------------------
# SPLIT BY YEAR AND EXPORT
# ---------------------------------------------------------
years = sorted(df["year"].unique())

print("\n📆 Years found:", years)
print("--------------------------------------------------")

for year in years:
    df_year = df[df["year"] == year]

    output_path = os.path.join(output_folder, f"Castello_{year}.csv")
    df_year.to_csv(output_path, index=False)

    print(f"✔ Saved {len(df_year)} rows for year {year} → {output_path}")

print("\n🎉 DONE: All yearly CSVs exported successfully.")
```
