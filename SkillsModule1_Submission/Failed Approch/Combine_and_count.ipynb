```python
import pandas as pd
import glob
import os

# 📁 Folder where all CSV files are stored
input_folder = r"C:\Users\TUF\Desktop\Architecture\UCL\Design\Public_Spaces_Datascraping\All_Sentiment\*"

# 🔍 Find all CSV paths
csv_files = glob.glob(input_folder)

print("Detected files:")
print(csv_files)

if not csv_files:
    print("⚠️ No CSV files found. Check file names inside All_Sentiment.")
    raise SystemExit

combined_data = []
file_counts = []

for file in csv_files:
    filename = os.path.basename(file)

    # Extract name before "__Sentiment_CLIP.csv"
    base_name = filename.replace("__Sentiment_CLIP.csv", "")

    # Read CSV
    df = pd.read_csv(file)

    # Count rows for summary report
    file_counts.append([base_name, len(df)])

    # Add new column to beginning
    df.insert(0, "location_name", base_name)

    combined_data.append(df)

# 🧩 Combine everything
combined_df = pd.concat(combined_data, ignore_index=True)

# 📤 Output files
combined_output = r"C:\Users\TUF\Desktop\Architecture\UCL\Design\Public_Spaces_Datascraping\1Combined\combined_all_locations.csv"
combined_df.to_csv(combined_output, index=False)

count_output = r"C:\Users\TUF\Desktop\Architecture\UCL\Design\Public_Spaces_Datascraping\1Combined\location_counts.csv"
pd.DataFrame(file_counts, columns=["location_name", "num_rows"]).to_csv(count_output, index=False)

print("Done!")
print("Combined CSV saved to:", combined_output)
print("Count CSV saved to:", count_output)
```
