```python
import pandas as pd
import requests
import cv2
import numpy as np
from io import BytesIO
from PIL import Image
from ultralytics import YOLO
from insightface.app import FaceAnalysis

# =============================================
# CONFIGURATION
# =============================================
INPUT_CSV = "Campo_San_Lorenzo_Indoor_Outdoor_Human.csv"
OUTPUT_CSV = "Campo_San_Lorenzo_with_AgeGroups.csv"

# =============================================
# LOAD MODELS
# =============================================
print("Loading YOLOv8 model...")
yolo = YOLO("yolov8n.pt")  # person detection

print("Loading InsightFace age estimator...")
fa = FaceAnalysis(name="buffalo_l", providers=["CPUExecutionProvider"])
fa.prepare(ctx_id=0, det_size=(640, 640))


# =============================================
# FUNCTIONS
# =============================================
def load_image_from_url(url):
    """Loads image from a URL into OpenCV format."""
    try:
        headers = {"User-Agent": "Mozilla/5.0"}
        response = requests.get(url, headers=headers, timeout=10)
        response.raise_for_status()
        img = Image.open(BytesIO(response.content)).convert("RGB")
        return cv2.cvtColor(np.array(img), cv2.COLOR_RGB2BGR)
    except Exception as e:
        print(f"Failed to load image: {url} | {e}")
        return None


def convert_age_to_group(age):
    """Convert estimated age into your categories."""
    if age < 0:
        return None
    if age <= 12:
        return "child"
    if age <= 19:
        return "teenager"
    if age <= 35:
        return "young_adult"
    if age <= 60:
        return "adult"
    return "elderly"


# =============================================
# MAIN PROCESS
# =============================================
def main():
    print("\nLoading CSV...")
    df = pd.read_csv(INPUT_CSV)

    if "url_s" not in df.columns:
        raise KeyError("CSV must contain a 'url_s' column.")

    age_results = []

    for idx, row in df.iterrows():
        url = row["url_s"]
        print(f"\nProcessing {idx+1}/{len(df)} → {url}")

        img = load_image_from_url(url)
        if img is None:
            age_results.append("image_load_error")
            continue

        # STEP 1: Detect people (bounding boxes)
        yolo_results = yolo(img, verbose=False)
        person_boxes = []
        for b in yolo_results[0].boxes:
            if int(b.cls[0]) == 0:  # class 0 = person
                x1, y1, x2, y2 = map(int, b.xyxy[0])
                person_boxes.append((x1, y1, x2, y2))

        # STEP 2: Face detection + age estimation using InsightFace
        faces = fa.get(img)
        img_age_groups = []

        for f in faces:
            age = f.age
            group = convert_age_to_group(age)
            if group:
                img_age_groups.append(group)

        age_results.append(img_age_groups)

    # Save to CSV
    df["age_groups"] = age_results
    df.to_csv(OUTPUT_CSV, index=False)

    print("\n===========================================")
    print("PROCESS COMPLETED!")
    print(f"Saved output CSV → {OUTPUT_CSV}")
    print("===========================================")


if __name__ == "__main__":
    main()
```
